```python
import pickle
from pprint import pprint
```

```python
with open("scan_results_data", "rb") as file:
    data = pickle.load(file)
```

```python
res = dict()
for r in data:
    ip, d = list(r.items())[0]
    if ip not in res.keys():
        res[ip] = list()
    res[ip].append(d)
    
```

```python
def chain(*iterables):
    # chain('ABC', 'DEF') --> A B C D E F
    for it in iterables:
        for element in it:
            yield element
            
            
def from_iterable(iterables):
    # chain.from_iterable(['ABC', 'DEF']) --> A B C D E F
    for it in iterables:
        for element in it:
            yield element
```

```python
for k in res.keys():
    res[k] = sorted([i for i in from_iterable(res[k])], key=lambda x:x['port'])
    
```

```python
for k in res.keys():
    res[k] = sorted(res[k], key=lambda x:x['port'])
    
```

```python
for i in res.keys():
    print(f"\nScanned results for {i}\n {'*'*60}")
    print("{:<10} {:<10} {:<10}".format('Ports', 'States', 'Service', 'Banner' ))
    for d in res[i]:
        print("{:<10} {:<10} {:<10}".format(d['port'], d['status'], 'Unknown', d['banner'] ))
```
